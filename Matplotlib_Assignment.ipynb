### Task 1
#### Scipy: We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset given below.

##### 1. fitting it to the periodic function
##### 2. plot the fit
##### Data
##### Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
##### Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18
##### Expected Output:


#### 1. Fitting it to the periodic function

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import optimize
%matplotlib inline
```

```python
max_temp = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]) 
min_temp = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
months= np.arange(12)
def func_temp(temp, avg, freq, offset):
    return (avg
            + freq * np.cos((temp + offset) * 1.8 * np.pi / temp.max()))

res_max, cov_max = optimize.curve_fit(func_temp, months,
                                      max_temp,[40,10,0])
res_min, cov_min = optimize.curve_fit(func_temp, months,
                                      min_temp,[-40, 20, 0])

```

#### Plot the Fit

```python

days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(months, max_temp, 'ro')
plt.plot(days, func_temp(days, *res_max), 'm-')
plt.plot(months, min_temp, 'bo')
plt.plot(days, func_temp(days, *res_min), 'y-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```
